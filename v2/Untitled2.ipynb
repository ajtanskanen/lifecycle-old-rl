```python
import numpy as np
from scipy.interpolate import interp1d

hajonta=np.zeros((100,3))
```

```python
hajonta[20,:]=[0.82707,1.02568,1.26636]
hajonta[25,:]=[0.82596,1.02855,1.34849 ]
hajonta[30,:]=[0.77539,1.03629,1.44658 ]
hajonta[50,:]=[0.729356592,1.053135396,1.698384385 ]
```

```python
def ifunc(x1,x2,hajonta,q):
    x = np.array([x1,x2])
    y = np.array([hajonta[x1,q],hajonta[x2,q]])
    f = interp1d(x, y)
    x2 = np.linspace(x1, x2, num=x2-x1, endpoint=True)
        
    return f(x2)
```

```python
for q in range(3):
    hajonta[20:25,q]=ifunc(20,25,hajonta,q)
    hajonta[25:30,q]=ifunc(25,30,hajonta,q)    
    hajonta[30:50,q]=ifunc(30,50,hajonta,q)
    hajonta[51:100,q]=hajonta[50,q]
```
